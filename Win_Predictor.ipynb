We need only 2nd innings data for win prediction

```python
ipl = ipl[ipl['innings']==2]
```

```python
# Merge the two datasets on the common columns
df = pd.merge(ipl, toss, left_on=['start_date','season','venue','batting_team','bowling_team'],
                     right_on=['Date','season','Venue','Team2','Team1'])
```

```python
#Removing season 2022 for testing the trained model later
df = df[df['season']!= 2022]
```

```python
df.head(4)
```

```python
df = df.drop(['Date','Venue','Match ID'], axis=1)
```

```python
df.head(3)
```

```python
df['Won by'].value_counts()
```

```python
# Calculating current run scored by team
df['runs'] = df['runs_off_bat'] + df['extras']
df['current_score'] = df.groupby(['match_id', 'batting_team'])['runs'].cumsum()
```

```python
# Calculating wickets lost by team
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if pd.isna(x) else 1)
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['wickets'] = df.groupby(['match_id','batting_team']).cumsum()['player_dismissed']
```

```python
df['over'] = df['ball'].apply(lambda x: str(x).split(".")[0]).astype('int')
df['over_ball'] = df['ball'].apply(lambda x: str(x).split(".")[1]).astype('int')
df['balls'] = (df['over'])*6 + df['over_ball']
df['crr'] = round((df['current_score'])/(df['over']+(df['over_ball']/6)),2)
```

```python
df['rrr'] = round((df['First innings score']- df['current_score'])/((126-df['balls'])/6),2)
```

```python
def result(row):
    return 1 if row['batting_team'] == row['Match winner'] else 0
df['result'] = df.apply(result,axis=1)
```

```python
df.head(4)
```

```python
df.info()
```

```python
final_df = df[['venue','batting_team','bowling_team','First innings score','crr','rrr','wickets','balls','result']]
```

```python
final_df.head(3)
```

```python
X = final_df.drop('result',axis = 1)
y = final_df['result']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print("X train shape is {} \nX_test shape is {}".format(X_train.shape,X_test.shape))
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['venue','batting_team','bowling_team'])
]
,remainder='passthrough')
```

# Model Building

As we are going to predict the probability of the outcome not the outcome of the match so we will use machine learning models based on probability like

1. Logistic Regression
2. Support Vector Machine(SVM)
3. Naive Baiye's
4. Random Forest
5. Gradient Boosting

```python
# Fit and transform X_train using trf
X_train_trans = trf.fit_transform(X_train)

# Transform X_test using trf
X_test_trans = trf.transform(X_test)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
lr = LogisticRegression(random_state=0,solver='liblinear')
lr.fit(X_train_trans,y_train)
y_lr = lr.predict(X_test_trans)
```

```python
accuracy_score(y_test,y_lr)
```

```python
#Naive Bayes
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train_trans,y_train)
y_nb = nb.predict(X_test_trans)
```

```python
accuracy_score(y_test,y_nb)
```

```python
#Random Forest 
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 12, random_state=0)
rfc.fit(X_train_trans,y_train)
y_rfc = rfc.predict(X_test_trans)
```

```python
accuracy_score(y_test,y_rfc)
```

```python
y_df = rfc.predict_proba(X_test_trans)
y_df[2]
```

```python
#Gradient Boosting
from sklearn.ensemble import GradientBoostingClassifier

# creating and training the model
gb = GradientBoostingClassifier(n_estimators= 50,learning_rate=0.05,max_depth= 12)
gb.fit(X_train_trans, y_train)

# making predictions on the test set
y_gb = gb.predict(X_test_trans)
```

```python
accuracy_score(y_test, y_gb)
```

```python
y_gb1 = gb.predict_proba(X_test_trans)
y_gb1[2]
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.pipeline import Pipeline
```

```python
model = Pipeline(steps=[
    ('step1',trf),
    ('step2',GradientBoostingClassifier(n_estimators= 50,learning_rate=0.05,max_depth= 12))
])
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
